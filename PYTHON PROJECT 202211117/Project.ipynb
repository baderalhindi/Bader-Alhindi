### For JOD To KWH

```python
def supported_val1(v1):
    if v1<=300*0.05:
        return v1/0.05
    elif v1<=(300*0.05+300*0.10):
        return 300+(v1-300*0.05)/0.10
    else:
        return 600 + (v1-(300*0.05+300*0.10))/0.20
def unsupported_val1(c1):
    if c1<=1000*0.12:
        return c1/0.12
    else:
        return 1000+(c1-1000*0.12)/0.15
```

### For Budget Display

```python
def Company(budget):
    com_l=[]
    for i,v in enumerate(df['JOD Per 1 KW']):
        if budget>=v:
            o=df.loc[i]
            com_l.append(o)
    return com_l
```

### Check If The Company Exist Or Not

```python
def Check_Company(aa):
    for i in df['Installation Company']:
        if(aa==i):
            return True
    return False    
```
